```python
import torch
import numpy as np
import matplotlib.pyplot as plt

def loss(x,y,w):
    y_pred = forward(x,w)
    return (y-y_pred)**2
    
def forward(x,w):
    return x*w

data = np.array([
    [1.0,2.0,3.0],
    [2.0,4.0,6.0]
]).T
x_data,y_data = data[:,:-1],data[:,-1]
x_data = torch.from_numpy(x_data)
y_data = torch.from_numpy(y_data)

w = torch.Tensor([1.0])
w.requires_grad = True

print("predict(before traning)",4,forward(4,w).item())

for epoch in range(100):
    for x,y in zip(x_data,y_data):
        l = loss(x,y,w)
        l.backward()
        print("\tgrad:",x,y,w.grad.item())
        w.data = w.data - 0.01*w.grad.data
        w.grad.data.zero_() # 清空梯度
    print("progress:",epoch,l.item())
print("predict (after training)",4,forward(4,w).item())
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt

def loss(x,y,w1,w2,b):
    y_pred = forward(x,w1,w2,b)
    return (y-y_pred)**2
    
def forward(x,w1,w2,b):
    return w1*(x**2)+w2*x+b

data = np.array([
    [1.0,2.0,3.0],
    [2.0,4.0,6.0]
]).T
x_data,y_data = data[:,:-1],data[:,-1]
x_data = torch.from_numpy(x_data)
y_data = torch.from_numpy(y_data)

w1 = torch.Tensor([1.0])
w2 = torch.Tensor([1.0])
b = torch.Tensor([1.0])
w1.requires_grad = True
w2.requires_grad = True
b.requires_grad = True

print("predict(before traning)",4,forward(4,w1,w2,b).item())

for epoch in range(10):
    for x,y in zip(x_data,y_data):
        l = loss(x,y,w1,w2,b)
        l.backward()
        print("\tgrad:",x,y,w1.grad.item(),w2.grad.item(),b.grad.item())
        w1.data = w1.data - 0.01*w1.grad.data
        w2.data = w2.data - 0.1*w2.grad.data
        b.data = b.data - 0.1*b.grad.data
        w1.grad.data.zero_()
        w2.grad.data.zero_()
        b.grad.data.zero_()
    print("progress:",epoch,l.item())
print("predict (after training)",4,forward(4,w1,w2,b).item(),w1.data.item(),w2.data.item(),b.data.item())
```
